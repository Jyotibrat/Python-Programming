```python
# Print the number is a prime number or not.
def is_prime(number):
    if number <= 1:
        return False
    if number == 2:
        return True
    if number % 2 == 0:
        return False
    for i in range(3, int(number**0.5) + 1, 2):
        if number % i == 0:
            return False
    return True

try:
    number = int(input("Enter a number: "))
    if is_prime(number):
        print(f"{number} is a prime number")
    else:
        print(f"{number} is not a prime number")
except ValueError:
    print("Please enter a valid integer.")
```

```python
# Print the matrix after doing Matrix multiplication.
import numpy as np

def matrix_multiplication(matrix1, matrix2):
    if matrix1.shape[1] != matrix2.shape[0]:
        raise ValueError("Number of columns in the first matrix must be equal to the number of rows in the second matrix.")
    return np.dot(matrix1, matrix2)

def main():
    matrix1 = np.array([[1, 2, 3], [4, 5, 6]])
    matrix3 = np.array([[1, 2], [3, 4], [5, 6]])

    try:
        result_multiplication = matrix_multiplication(matrix1, matrix3)
        print("Matrix Multiplication Result:")
        print(result_multiplication)
    except ValueError as e:
        print(f"Multiplication Error: {e}")

if __name__ == "__main__":
    main()
```

```python
import numpy as np

def matrix_addition(matrix1, matrix2):
    if matrix1.shape != matrix2.shape:
        raise ValueError("Matrices must have the same dimensions for addition.")
    return matrix1 + matrix2

def main():
    matrix1 = np.array([[1, 2, 3], [4, 5, 6]])
    matrix2 = np.array([[7, 8, 9], [10, 11, 12]])

    try:
        result_addition = matrix_addition(matrix1, matrix2)
        print("Matrix Addition Result:")
        print(result_addition)
    except ValueError as e:
        print(f"Addition Error: {e}")

if __name__ == "__main__":
    main()
```
