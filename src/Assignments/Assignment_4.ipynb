```python
#Write a program in python to check whether a given string is palindrome or not.

string = input("Enter a string: ")

if string == string[::-1]: print("The entered string is a palindrome string.")
else: print("The entered string is a palindrome string.")
```

```python
"""
Write a Python function that accepts a string and calculate the number of upper case letters
and lower case letters.
Sample String : &#39;The quick Brow Fox&#39;
Expected Output :
No. of Upper case characters : 3
No. of Lower case Characters : 12
"""

def check_case(string):
  upper_case = int(0)
  lower_case = int(0)
  for characters in string:
    if characters.isupper():
      upper_case += 1
    elif characters.islower():
      lower_case += 1
  return upper_case, lower_case

u_case, l_case = check_case(input("Enter a string: "))
print(f"No. of Upper case characters : {u_case}\nNo. of Lower case Characters : {l_case}")
```

```python
"""
Write a Python function to check whether a string is pangram or not. Note : Pangrams are
words or sentences containing every letter of the alphabet at least once.
For example : &quot;The quick brown fox jumps over the lazy dog&quot;
"""

def check_pangram(string):
  alphabet = "abcdefghijklmnopqrstuvwxyz"
  for characters in alphabet:
    if characters not in string.lower():
      return False
  return True

if check_pangram(input("Enter a string: ")): print("The entered string is a pangram.")
else: print("The entered string is not a pangram.")
```

```python
#Write a python program to count the occurrence(frequency) of a particular character in a string.

string = input("Enter a string: ")
character = input("Enter the character whose frequency you want to count: ")

count = int(0)
for characters in string:
  if characters == character:
    count += 1
print(f"The frequency of {character} in {string} is {count}")
```

```python
"""
Q6:Write a Python program to get a string made of the first 2 and the last 2 chars from a given
a string. If the string length is less than 2, return instead of the empty string. Sample String :
&#39;wakawama&#39;
Expected Result : &#39;wama&#39;
Sample String : &#39;wa&#39;
Expected Result : &#39;wawa&#39;
Sample String : &#39; w&#39;
Expected Result : Empty String
"""

string = input("Enter a string: ")
if len(string) < 2:
  print("Empty String")
else:
  last_two_characters = string[-2:]
  first_two_characters = string[0:2]
  print(f"Expected Result: {first_two_characters + last_two_characters}")
```

```python
"""
Write a Python program to find the first appearance of the substring &#39;not&#39; and &#39;poor&#39; from a
given string, if &#39;bad&#39; follows the &#39;poor&#39;, replace the whole &#39;not&#39;...&#39;poor&#39; substring with &#39;good&#39;.
Return the resulting string.
Sample String : &#39;The lyrics is not that poor!&#39;
Expected Result : &#39;The lyrics is good!&#39;
"""

def replace_not_poor_with_good(s):
    not_index = s.find('not')
    poor_index = s.find('poor')

    if not_index != -1 and poor_index != -1 and poor_index > not_index:
        s = s[:not_index] + 'good' + s[poor_index+4:]

    return s

string = input("Enter a string: ")
result = replace_not_poor_with_good(string)
print(f"Expected Result: {result}")
```

```python
"""
Write a Python program to get a string from a given string where all occurrences of its first
char have been changed to &#39;$&#39;, except the first char itself.
Sample String : &#39;restart&#39;
Expected Result : &#39;resta$t’
"""

string = input("Enter a string: ")
first_char = string[0]
New_string = string[1:]
for characters in New_string:
  if characters == first_char:
    New_string = New_string.replace(characters, "$")
print(f"Expected Result: {first_char + New_string}")
```
