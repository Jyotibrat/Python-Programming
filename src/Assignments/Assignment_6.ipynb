```python
n = int(input("Enter a number: "))

new_n = 0
while n > 0:
    digit = n % 10
    new_n = (new_n * 10) + digit
    n //= 10

print(f"Reversed Number: {new_n}")
```

```python
P = 4
Q = 2

first_term = int((2+(8*P))/2)

second_term = int(((P-Q)*(P+Q))/2)

third_term = int(((P+Q)/2)*4)
print(f"The required value is: {first_term + second_term + third_term}")
```

```python
Messages = {
    "warm and dry": "Play Basketball",
    "warm and Humid": "Play Tennis",
    "cold and dry": "Play Cricket",
    "cold and humid": "Swim"
}

temp = input("Enter the temperature: ").lower().strip()
hum = input("Enter the humidity: ").lower().strip()

message = temp + " and " + hum

if message in Messages:
    print(Messages[message])
else:
  print("Do not go outside.")
```
