```python
#Write a program to input two numbers from the user and check its an integer or not using try and except

n = input("Enter a number: ")
try:
    int(n)
    print("It is an integer")
except:
    print("It is not an integer")
```

```python
#Write a program to input a string from the user evaluate using try and except

try:
    s = int(input("Enter a string: "))
    print("It is not a string.")
except:
    print("It is a string.")
```

```python
#Write a program to input two number from the user and check its an integer or not using try and except, if number are integer then the print sum of that number using finally block

try:
  n1 = input("Enter first number: ")
  n2 = input("Enter second number: ")
  num1 = int(n1)
  num2 = int(n2)
except:
  print("The entered numbers are not integers.")
finally:
    print(f"The sum of the entered numbers is: {num1 + num2}")
```

```python
#Write a program to input two number from the user and handle your exception using else.

n1 = input("Enter first number: ")
n2 = input("Enter second number: ")

if n1.isdigit() and n2.isdigit():
  print("The entered numbers are integers.")
else:
  print("The entered numbers are not integer.")
```

```python
#Write a program to print multiple except block.

def multiple_excepts():
    try:
        number = int(input("Enter a number: "))
        result = 10 / number
        print(f"Result is {result}")

        my_list = [1, 2, 3]
        print(my_list[number])

    except ValueError:
        print("ValueError: Please enter a valid integer.")

    except ZeroDivisionError:
        print("ZeroDivisionError: Division by zero is not allowed.")

    except IndexError:
        print("IndexError: Index out of range. The list has fewer elements.")

multiple_excepts()
```
