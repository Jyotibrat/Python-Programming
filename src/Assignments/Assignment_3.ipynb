```python
#Write a Python function to find the Max of three numbers.

def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

x, y, z = tuple(map(int, input("Enter three numbers").split()))
print(f"The largest of {x}, {y} and {z} is: {find_max(x, y, z)}")
```

```python
#Write a Python function to sum all the numbers in a list.

def sum_list(initial_list):
  sum = int(0)
  for i in initial_list:
    sum += i
  return sum

print("The sum of the list of numbers:",sum_list(list(map(int, input("Enter a list of numbers").split()))))
```

```python
#Write a Python function to multiply all the numbers in a list.

def multiply_list(initial_list):
  multiply = int(0)
  for i in initial_list:
    multiply += i
  return multiply

print("The multiplication of the list of numbers:",multiply_list(list(map(int, input("Enter a list of numbers").split()))))
```

```python
#Write a Python function to reverse a string.

def reverse_string(word):
  return word[::-1]

print("The Reverse of the String is:", reverse_string(input("Enter a String: ")))
```

```python
#Write a Python function to calculate the factorial of a number (non-negative integer). The function accept the number as an argument

def fact(n):
  f = int(1)
  if n < 0:
    return "Factorial is not defined for negative numbers"
  elif n == 0 or n == 1:
    return 1
  else:
    for i in range(1, n + 1):
      f *= i
    return f

print("The factorial of the number is:", fact(int(input("Enter a number: "))))
```

```python
#Write a Python function to check whether a number is in a given range.

def check(n, first, last):
  if first <= n <= last:
    return True
  else:
    False

if check(int(input("Enter a number: ")), int(input("Enter the first number of the range: ")), int(input("Enter the last number of the range: "))): print("The entered number is in the given range.")
else: print("The entered number is not in the given range.")
```

```python
#Write a Python function that takes a number as a parameter and check the number is prime or not.

def check_prime(n):
  count = int(0)
  for i in range(1, n + 1):
    if n % i == 0:
      count += 1
  if count == 2:
    return True
  else:
    return False

if check_prime(int(input("Enter a number: "))): print("The entered number is a prime number.")
else: print("The entered number is not a prime number.")
```

```python
#Write a Python function that checks whether a passed number is palindrome or not.

def check_palindrome(n):
  return str(n) == str(n)[::-1]

if check_palindrome(int(input("Enter a number: "))): print("The entered number is palindrome number.")
else: print("The entered number is not a palindrome number.")
```
