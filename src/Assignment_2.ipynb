```python
"""
1. Write a Python program to count the number of strings from a given list of
strings. The string length is 2 or more and the first and last characters are the same.
Sample List : [abc, xyz, aba, 1221]
"""

strings = ['abc', 'xyz', 'aba', '1221']
count = int(0)
for i in strings:
  if len(i) >= 2:
    if i[0] == i[-1]:
      count += 1

print(count)
```

```python
"""
2. Write a Python program to get a list, sorted in increasing order by the last
element in each tuple from a given list of non-empty tuples.
Sample List : [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
"""

initial_list = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
final_list = [initial_list[0]]
initial_list.pop(0)
for i in initial_list:
  for j in final_list:
    if i[1] < j[1]:
      final_list.insert(final_list.index(j), i)
      break

print(final_list)
```

```python
"""
3. Write a Python program to find the list of words that are longer than n from a
given list of words.
"""

n = int(input("Enter a number: "))
print("Enter the input list numbers with spaces between: ", end="")
input_list = list(input().split())
final_list = []
for i in input_list:
  if len(i) > n:
    final_list.append(i)
print(final_list)
```

```python
"""
4. Write a Python function that takes two lists and returns True if they have at least
one common member.
"""

def take_input(list_1, list_2):
  flag = False
  for i in list_1:
    for j in list_2:
      if i == j:
        flag = True
        break
  return flag

print("Enter the elements of the first list: ", end="")
list_1 = list(map(int, input().split()))

print("Enter the elements of the second list: ", end="")
list_2 = list(map(int, input().split()))

print(take_input(list_1, list_2))
```

```python
"""
5. Write a Python program to print a specified list after removing the 0th, 4th and
5th elements.
Sample List : [&#39;Red&#39;, &#39;Green&#39;, &#39;White&#39;, &#39;Black&#39;, &#39;Pink&#39;, &#39;Yellow&#39;]
"""

initial_list = ['Blue', 'Red', 'Green', 'White', 'Brown', 'Violet', 'Black', 'Pink', 'Yellow']
final_list = []
for i in initial_list:
  if initial_list.index(i) != 0 and initial_list.index(i) != 4 and initial_list.index(i) != 5:
    final_list.append(i)

print(final_list)
```
