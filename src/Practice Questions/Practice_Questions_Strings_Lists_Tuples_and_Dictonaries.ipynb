```python
string = input("Enter a string: ")
count_for_uppercase = 0
count_for_lowercase = 0
for character in string:
  if character.isupper():
    count_for_uppercase += 1
  elif character.islower():
    count_for_lowercase += 1
print("Number of uppercase characters:", count_for_uppercase)
print("Number of lowercase characters:", count_for_lowercase)
```

```python
def getVowels(word):
  vowels = 'aeiou'
  list_of_vowels = []

  for char in word:
    if char in vowels:
      list_of_vowels.append(char)
  return list_of_vowels

w = input("Enter a word: ")
print(getVowels(w))
```

```python
word_1 = input("Enter the first word: ")
word_2 = input("Enter the second word: ")

list_of_common_chars = []

for char in word_1:
  if char in word_2 and char not in list_of_common_chars:
    list_of_common_chars.append(char)

print("Common characters:", list_of_common_chars)
```

```python
n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))
print(f"Bitwise AND operation: {n1&n2}")
print(f"Bitwise OR operation: {n1|n2}")
print(f"Bitwise XOR operation: {n1^n2}")
print(f"Bitwise NOT operation: {~n1}")
print(f"Left shift operation: {n1<<2}")
print(f"Right shift operation: {n1>>2}")
```

```python
from collections import ChainMap

dict_1 = {"Name": "Akshit", "Age": 18}
dict_2 = {"Reg. No.": "23BAI11234"}

chain_map = ChainMap(dict_1, dict_2)

print(chain_map["Name"])
print(chain_map["Age"])
print(chain_map["Reg. No."])
print(chain_map.parents)
print(chain_map.new_child({"course": "CSE AI & ML"}))
print(chain_map.maps)
```
