```python
import random

l = []
count = 0
for i in range(1, 1001):
  n = random.randint(1, 1000)
  if n % 5 == 0 and n % 7 == 0:
    if n not in l:
      l.append(n)
      count+=1
  if count == 5:
    break
print(l)
```

```python
string = input("Enter a string: ")
new_string = ""

for index, character in enumerate(string):
    if index % 2 != 0:
        new_string += character

print(new_string)
```
