```python
s = "This is my file handling program"
file = open("demo1.txt", "w")
file.write(s)
print("file created")
file.close()
```

```python
file = open("demo1.txt", "r")
filecontent = file.read()
print(filecontent)
```

```python
l1 = ["Python", "Java", "php", "Angular.js"]
file = open("demo2.txt", "w")
file.writelines(l1)
print("file created")
file.close()
```

```python
file = open("demo2.txt")
filelist = file.readlines()
print(filelist)
```

```python
file = open("demo2.txt", "a")
file.write("C++")
print("file updated")
file.close
```

```python
try:
  f = open("C:\myfile.txt", "r")
except FileNotFoundError as msg:
  print("File not found.")
else:
  print("Data from file.")
  print(f.read())
finally:
  print("I am going to close the file.")
  f.close()
```
