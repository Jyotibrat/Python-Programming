```python
# Write a program to implement both overloading and overiding.

class Parent():

  def First_function(self, a = None, b = None, *args):

    if a is not None and b is not None:
      return(a+b)
    elif args:
      return(sum(args))

class Child(Parent):

  def Second_function(self):

    return super().First_function(1, 2)

obj1 = Parent()
obj2 = Child()
print(obj1.First_function(4, 4, 4))
print(obj1.First_function(2, 3))
print(obj2.Second_function())
```
