```python
#Method Overloading.

class Vehicle():
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Move!")

class Car(Vehicle):
  pass

class Boat(Vehicle):
  def move(self):
    print("sail!")

class Plane(Vehicle):
  def move(self):
    print("fly!")

car1 = Car("Ford", "Mustang")
boat1 = Boat("Ibiza", "Touring 20")
plane1 = Plane("Boeing", "747")

for x in (car1, boat1, plane1):
  print(x.brand)
  print(x.model)
  x.move()
```

```python
#Method Overloading.

class Area():
  def find(self, a=None, b=None):
    if a!=None and b!=None:
      print(f"Area of Rectangle: {a*b}")
    elif a!=None:
      print(f"Area of square: {a*a}")
    else:
      print("Nothing to find")

obj1 = Area()
obj1.find()
obj1.find(10)
obj1.find(10,20)
```

```python
class Ws:
    def display_info(self):
        print("Welcome to VIT")

class ITP(Ws):
    def display_info(self):
        display_info()
        print("Welcome to ITP class")

obj1 = ITP()
obj1.display_info()
```
