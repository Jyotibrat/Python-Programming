```python
# Open a file in read mode
file_path = 'example.txt'  # Replace with the path to your file
try:
    with open(file_path, 'r') as file:
        # Read the first 10 characters
        content = file.read(10)
        print(f"First 10 characters: {content}")

        # Use seek to move back to the start of the file
        file.seek(0)
        position = file.tell()
        print(f"Current position after seek to start: {position}")

        # Read the next 5 characters
        content = file.read(5)
        print(f"Next 5 characters: {content}")

        # Use seek to move to the 20th byte in the file
        file.seek(20)
        position = file.tell()
        print(f"Current position after seek to 20th byte: {position}")

        # Read the next 10 characters from the new position
        content = file.read(10)
        print(f"10 characters from 20th byte: {content}")

except FileNotFoundError:
    print(f"The file {file_path} does not exist.")
except Exception as e:
    print(f"An error occurred: {e}")
```
