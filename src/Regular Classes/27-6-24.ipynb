```python
class Democlass:
  a = 20
  def __init__(self):
    print("Welcome to VIT")
  def showvalue(self):
    self.c = self.a * self.a
    print(self.c)

  def showvalue1(self, a, b):
    print(a + b)

obj = Democlass()
obj.showvalue()
obj.showvalue1(10, 20)
```

```python
class faculty:
  def putdata(self):
    self.id = int(input("Enter faculty id: "))
    self.name = input("Enter faculty name: ")
    self.salary = int(input("Enter faculty salary: "))

  def display(self):
    print(f"Faculty id: {self.id}")
    print(f"Faculty name: {self.name}")
    print(f"Faculty salary: {self.salary}")

a = faculty()
a.putdata()
a.display()
```
